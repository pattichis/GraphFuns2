# Make a beautiful flower

1. Make the first pedal look nice by reducing the domain.
2. Create the other pedals. Have fun coloring your pedals.
3. After you get the pedals, set hide_axes=True in graph_funs().
4. Remove plot.plot_funs() function calls to plot one pedal at a time only.
5. Save each pedal into a different picture.

Here is a list of colors:
   "blue", "green", "red", "cyan", "magenta", "yellow", "black", "white","olive","indigo".

```python
# MODIFY to add multiple transformations

# Define x as a symbolic variable
x = sp.symbols('x')

# Define exponential function
f1 = 0.25*(2)**(x)

# Create a graph object with equal_axes)
plot = graph_funs(equal_axes=True, hide_axes=False)

# Setup the grid
# Determine the range for X based on reflections and translations.
# Determine the range for Y based on reflections and translations (look at maximum values).
plot.set_XY(minX=-10, maxX=+10, minY=-10, maxY=+10)

# Add to the list of functions:
domain1 = [0.1, 5, 20]

# Plot fun
plot.add_fun(f1, domain1, 'f1(x)='+str(f1), 'red')  # Add the original function.
plot.plot_funs(plot_title="Original function f1 (red)", img_name="frame1.png")

f2 = x
domain2 = [-10, 10, 10]
plot.add_fun(f2, domain2, 'f2', 'black')
plot.plot_funs(plot_title="f1, f2", img_name="frame2.png")

# Add the inverse function.
domain3 = domain1
f3 = plot.add_inverse_fun(f1, domain3, 'inverse of f1', 'green')
plot.plot_funs(plot_title="f1, f2, inverse f1", img_name="frame3.png")

# Flip the inverse function:
f4 = -f3
domain4 = domain1
plot.add_fun(f4, domain4, 'f4', 'blue')
plot.plot_funs(plot_title="f1, f2, f3, f4", img_name="frame4.png")

# Reflect f4:
f5 = f4.subs(x, -x)
domain5 = [-5, -0.1, 20]
plot.add_fun(f5, domain5, 'f5', 'blue')
plot.plot_funs(plot_title="f1, f2, f3, f4, f5", img_name="frame4.png")

# Reflect f5:
f6 = -f5
domain6 = domain5
plot.add_fun(f6, domain6, 'f6', 'blue')
plot.plot_funs(plot_title="f1, f2, f3, f4, f5, f6", img_name="frame4.png")

# Add y=x reflection.
f7 = -x
domain7 = [-10, 10, 20]
plot.add_fun(f7, domain7, 'f7', 'black')
plot.plot_funs(plot_title="f1-f7", img_name="frame4.png")
```
