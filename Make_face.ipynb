# Make a closed figure function
This the main function for plotting.

Here is a list of colors:
   "blue", "green", "red", "cyan", "magenta", "yellow", "black", "white","olive","indigo".

````python
def make_closed_fig(plot, a, b, h, k, d, num_of_points):
  """ Make a closed figure using:
       f(x)=a*b**(x-h)+k
  - a is a constant.
  - b is the base.
  - h is horizontal translation in (x-h).
  - k is vertical translation.
  Here, d is used to determine the domain: from -d to +d.
  num_of_points is the number of points to plot.
  plot specifies the plot object to plot on.
  """
  # Create the symbol
  x = sp.symbols('x')
  f0 = a*(b)**(x)+k

  # Specify mother function
  # Shift the domain by +h.
  f1 = f0.subs(x, x-h)
  domain1 = [0+h, d+h, num_of_points]

  # Invert it and move it up
  # Same domain.
  # Move f2 up
  f2 = -f1 + 2*f1.subs(x, d+h)
  domain2 = domain1

  # Reflect both:
  f3 = f0.subs(x, -x+h)
  domain3 = [h-d, 0+h, num_of_points]

  f4 = -f0.subs(x, -x+h) + 2*f3.subs(x, -d+h)
  domain4 = domain3

  plot.add_fun(f1, domain1, 'f1(x)='+str(f1), 'red')
  plot.add_fun(f2, domain2, 'f2(x)='+str(f2), 'blue')
  plot.add_fun(f3, domain3, 'f3(x)='+str(f3), 'green')
  plot.add_fun(f4, domain4, 'f4(x)='+str(f4), 'orange')
````

# Make a video of drawing a face
The mother function is given by:
$$ f(x)=a\cdot b^{(x-h)}+k $$
where:  
  - a is a constant.
  - b is the base.
  - h is horizontal translation in (x-h).
  - k is vertical translation.

Based on the flower assignment, we learned how to reflect
and connect different copies of the mother function.

Here, we use four copies to create a closed figure:
```python
  make_closed_fig(plot, a, b, h, k, d, num_of_points)
```
The function takes the parameters of the mother function and also the following parameters:
  - d specifies the domain: from -d to +d.
  - num_of_points is the number of points to plot.
  - plot specifies the plot object to plot on.


**Modify the code below to make your own version of the face. Recall:**
$$ f(x)=a\cdot b^{(x-h)}+k $$

````python
# Playing with b

# Create a graph object with equal_axes)
plot = graph_funs(equal_axes=True, hide_axes=False)

# Setup the grid
# Determine the range for X based on reflections and translations.
# Determine the range for Y based on reflections and translations (look at maximum values).
plot.set_XY(minX=-10, maxX=+10, minY=0, maxY=+40)

# Draw different parts of the face and save in different frames

# Frame 1: draw the face:
b=4
make_closed_fig(plot, a=1, b=b, h=0, k=0, d=2, num_of_points=20)
plot.plot_funs(plot_title="Drawing the face", img_name="frame1.png")

# Frame 2: Draw the eyes:
make_closed_fig(plot, a=0.5, b=b, h=5, k=0, d=2, num_of_points=20)
make_closed_fig(plot, a=0.5, b=b, h=10, k=0, d=2, num_of_points=20)
plot.plot_funs(plot_title="Drawing the ...", img_name="frame2.png")
````
